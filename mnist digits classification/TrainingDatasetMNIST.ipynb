```python
from keras.datasets import mnist
from google.colab.patches import cv2_imshow
import cv2
import matplotlib.pyplot as plt
from keras import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from tensorflow.keras.utils import to_categorical
import numpy as np
import tensorflow.keras
from sklearn.model_selection import train_test_split
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
a = np.append(x_train, x_test, 0)
```

```python
a.shape
```

```python
b = np.append(y_train, y_test, 0)
```

```python
b.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(a, b, test_size=0.4)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
qty = 10 * [0]
for i in range(x_train.shape[0]):
  for j in range(10):
    if y_train[i] == j:
      qty[j] += 1
print(qty)
```

```python
for i in range(100):
  if y_train[i] == 9:
    print(x_train[i].shape) 
    cv2_imshow(x_train[i])
    print(x_train[i])
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

y_test[12]
```

```python
#create Model cnn
cnn = Sequential()

#add layers
cnn.add(Conv2D(64, kernel_size=(3,3), activation='relu', input_shape=(28,28,1), padding='same'))
cnn.add(MaxPooling2D(pool_size=(2,2), strides=2))
cnn.add(Dropout(0.4))
cnn.add(Conv2D(32, kernel_size=(3,3), activation='relu', padding='same'))
cnn.add(MaxPooling2D(pool_size=(2,2), strides=2))
cnn.add(Dropout(0.4))
cnn.add(Flatten())
cnn.add(Dropout(0.4))
cnn.add(Dense(128, activation='relu'))
cnn.add(Dense(10, activation='softmax'))
```

```python
#compile
cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
cnn.summary()
```

```python
a =  cnn.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=50)
# cnn.save_weights('model.h5')
```
