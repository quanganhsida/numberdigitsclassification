```python
from PIL import Image, ImageOps
import PIL.ImageOps
import tensorflow.keras
from tensorflow.keras import models
import numpy as np
import glob 
import os, os.path
from google.colab.patches import cv2_imshow
import cv2 as cv
```

```python
path = '/content/drive/MyDrive/Number Classification /mnist digits classification/New folder'
precision_dict = {}
load_model = models.load_model('/content/drive/MyDrive/Number Classification /mnist digits classification/Saved_models')
for folder in glob.glob(path + '/*'):
  satisfied_case = 0
  total_case_in_folder = len([name for name in os.listdir(folder) if os.path.isfile(os.path.join(folder,name))])
  index = folder.rfind('/')
  label = folder[index + 1:]
  for img in glob.glob(folder + '/*.jpg'):
    image = cv.imread(img, cv.IMREAD_GRAYSCALE)
    resized_img = cv.resize(image, (28,28), interpolation = cv.INTER_AREA)
    image1 = cv.medianBlur(resized_img, 1)
    ret,th1 = cv.threshold(image1,127,255,cv.THRESH_BINARY_INV)
    cv2_imshow(th1)
    x = np.expand_dims(th1, axis=0)
    result = load_model.predict(x)
    max_result = np.argmax(result)
    print(max_result)
    print('\n')
    if str(np.argmax(result)) == str(label):
      satisfied_case += 1
  precision_dict[label] = round(int(satisfied_case) / int(total_case_in_folder), 2)
print('Accuracy rate of each class: ')
for numeric_class in precision_dict:
  print(numeric_class + ': ' + str(precision_dict[numeric_class]))
```
